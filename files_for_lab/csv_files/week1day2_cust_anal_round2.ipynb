    Given the contribution of the missing rows in the respective column labels to the customer lifetiem value of > 5 % or the contribution to the total number of records likethe car type ( 50%), and given we have no null client ID data ( client can be identified) and all Client life contribution data are inplace, it is deemed better NOT TO DROP ANY ROWS.THE NULL VALUES ARE FILLED UP WITH TBD  where it makes sense to reflect capturing of the data in categorical columns in case data is made availabe in the future. Null values in 'months since last claim' will use the mean of the data and'number of open complaints' will be set to 0. This will allow to calculate performance of a state for instance without biasing the data. Response null values will be set to '0'(0 as a string to avoid mixing data types) for easy comparison and to run the model without failures. we can attempt to populate the missing vaues for state by multiple loops in ranges equivalent to the weights of California, oregon and Arizona given they make up most of the date. However for now, the null values will be set to TO BE DETERMINED 'TBD' given the client ID is available. 

```python
df_mca['state'] = df_mca['state'].fillna('TBD')
```

```python
df_mca['response'] = df_mca['response'].fillna('0')
```

```python
import numpy as np
mean_months_since_last_claim = df_mca['months since last claim'].mean()
df_mca['months since last claim'] = df_mca['months since last claim'].fillna(mean_months_since_last_claim)
```

```python
df_mca['number of open complaints'] = df_mca['number of open complaints'].fillna(0)
```

```python
df_mca['vehicle class'] = df_mca['vehicle class'].fillna('TBD')
```
